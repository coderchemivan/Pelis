```python

import pandas as pd
import numpy as np
import json
from pandas.io.json import json_normalize
```

```python
file_ = input('Introduce el nombre del archivo: ')
with open('files/{file}.json'.format(file=file_), encoding='utf-8') as file:
    df = pd.read_json(file).astype(str)


df.drop_duplicates(keep='first',ignore_index=True,inplace=True)

df = df[~((df.rating == "['ND']") & (df.year == "['ND']") &
    (df.budget == "['ND']") & (df.boxOffice_collection == "['ND']") &
    (df.cast == "nan") & (df.personajes == "nan"))
    ]

df.to_csv('files/{file}.csv'.format(file = file_),index=False)
df
#df.dropna(subset=['genre','rati=ng','year','duracion','busget','boxOffice_collection'],inplace=False)
```
